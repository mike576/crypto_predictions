```python
from tweepy import Stream
from tweepy.streaming import StreamListener
 
class MyListener(StreamListener):
    def __init__(self):
        self.path = "init.json"

    def set_path(self,mypath):
        self.path = mypath
        
    def on_data(self, data):
        try:
            with open(self.path, 'a') as f:
                f.write(data)
                return True
        except BaseException as e:
            print("Error on_data: %s" % str(e))
        return True
 
    def on_error(self, status):
        print(status)
        return True
 


```

```python

from threading import Thread
from time import sleep

def add_listener(trackfilter,path):
    print('adding ',trackfilter)
    mylistener=MyListener()
    mylistener.set_path(path)
    twitter_stream = Stream(auth,mylistener)
    twitter_stream.filter(track=[trackfilter])
 

if __name__ == "__main__":
    thread = Thread(target = add_listener, args = ("#NEO","neo_stream.json", ))
    thread.start()
    sleep(1)
    thread = Thread(target = add_listener, args = ("#Verge","verge_stream.json", ))
    thread.start()
    sleep(1)
    thread = Thread(target = add_listener, args = ("#Bitcoin","bitcoin_stream.json", ))
    thread.start()
    sleep(1)
    thread = Thread(target = add_listener, args = ("#Dash","dash_stream.json", ))
    thread.start()
    sleep(1)
    thread = Thread(target = add_listener, args = ("#Tron","tron_stream.json", ))
    thread.start()
    sleep(1)
    thread = Thread(target = add_listener, args = ("#Lota","lota_stream.json", ))
    thread.start()
    sleep(1)
    thread.join()



```
